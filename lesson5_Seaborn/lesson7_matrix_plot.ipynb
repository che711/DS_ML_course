```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import scipy
```

```python
df = pd.read_csv('../course_data/05-Seaborn/country_table.csv')
df.head()
```

```python
 df=df.set_index('Countries')
df
```

```python
sns.heatmap(data=df.drop('Life expectancy',axis=1));
```

```python
sns.heatmap(data=df.drop('Life expectancy',axis=1), linewidth=0.5);
```

```python
sns.heatmap(data=df.drop('Life expectancy',axis=1), linewidth=0.5, annot=True);
```

```python
plt.figure(dpi=200)
sns.heatmap(data=df.drop('Life expectancy',axis=1), linewidth=0.5, annot=True);
```

```python
plt.figure(dpi=200)
sns.heatmap(data=df.drop('Life expectancy',axis=1), linewidth=0.5, annot=True, 
            cmap='viridis');
```

```python
plt.figure(dpi=200)
sns.heatmap(data=df.drop('Life expectancy',axis=1), linewidth=0.5, annot=True, 
            cmap='viridis', center=40);
```

```python
plt.figure(dpi=200)
sns.clustermap(data=df.drop('Life expectancy',axis=1), linewidth=0.5, annot=True, 
            cmap='viridis')
```
