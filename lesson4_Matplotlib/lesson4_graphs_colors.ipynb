```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#2605ff', label='x vs x') # hex code also supports
ax.plot(x,x+1,color='purple', label='x vs x+1') # hex code also supports
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,color='yellow', lw=5, ls='-.') # толщина линии (по умолчанию =1)
ax.plot(x,x+1,color='red', linewidth=1.5, linestyle='--') 
# возможноые варианты для параметра linestyle (ls): '--', '-', '-.',':', "steps"
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# возможноые варианты для параметра linestyle (ls): '--', '-', '-.',':', "steps"

lines= ax.plot(x,x**1.5, color='purple', lw=5)
lines[0].set_dashes([5,2,10,2,3,5])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# возможноые варианты для параметра linestyle (ls): '--', '-', '-.',':', "steps"

ax.plot(x,x**1.5, color='purple', lw=2, marker='o', ls='--')

ax.plot(x,x**1, color='black', lw=2, marker='o', markersize=10, ls='--', 
        markerfacecolor='red', markeredgewidth=4, markeredgecolor='orange')
```
