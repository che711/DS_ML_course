```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
a
```

```python
b = a**4
b
```

```python
x = np.arange(0,10)
x
```

```python
y = 2*x
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

# ставя ; в конце строки "axes.plot(x,y)" мы убираем обозначение графика -> [<matplotlib.lines.Line2D at 0x78c0402e8a70>]
# или можно использовать plt.show(axes.plot(x,y))
axes.plot(x,y);
#plt.show(axes.plot(x,y))
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

# ставя ; в конце строки "axes.plot(x,y)" мы убираем обозначение графика -> [<matplotlib.lines.Line2D at 0x78c0402e8a70>]
# или можно использовать plt.show(axes.plot(x,y))
axes.plot(a,b);
#plt.show(axes.plot(x,y))
```

```python
fig = plt.figure()

# Набор больших осей
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);

# Маленькие оси
axes2 = fig.add_axes([0.2,0.5, 0.25,0.25])
axes2.plot(x,y)
plt.show()

#plt.show(axes.plot(x,y))
```

```python
fig = plt.figure()

# Набор больших осей
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000) 
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Возведение в степень 4')

# Маленькие оси
axes2 = fig.add_axes([1.2,0.5, 0.25,0.25])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50) 
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Увеличенный фрагмент')

plt.show()

#plt.show(axes.plot(x,y))
```

```python
fig = plt.figure(figsize=(2,2), dpi=200)
axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

fig.savefig("./new_figure.png", bbox_inches='tight')
```
